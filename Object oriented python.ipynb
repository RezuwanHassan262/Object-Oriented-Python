```python
def zeros(n):
  c = 0
  l = len(n) - 1
  n = n[1:]
  for i in n:
    if int(i) == 0:
      c += 1
  if c == l:
    return("Yes")
  else:
    return("No")

def allSame(n):
  l = len(n)
  x = n[0]
  c = 0
  for i in n:
    if i == x:
      c+=1
  if c == l:
    return("Yes")
  else:
    return("No")


def consecuitive(n):
  l = []
  k = n[:-1]
  l = len(n)
  c = 0
  for i in k:
    x = int(i)
    p = n.index(i)
    if x == 0 and int(n[p+1]) == 9:
      c +=1
    elif x == 9 and int(n[p+1]) == 0:
      c +=1
    elif x+1 == int(n[p+1]):
      c += 1
    elif x-1 == int(n[p+1]):
      c += 1

  if c == l-1:
    return("Yes")
  else:
    return("No")



def palindrome(n):
  l1 = len(n)
  l2 = int(l1/2)
  c = 0
  if l1%2 ==0:
    for i in range(l2):
      if n[i] == n[((-1)*i)-1]:
        c += 1
    if c == l2:
      return("Yes")
    else:
      return("No")
  else:
    l2 = int((l1-1)/2)
    for i in range(l2):
      if n[i] == n[((-1)*i)-1]:
        c += 1
    if c == l2:
      return("Yes")
    else:
      return("No")

def unique(s):
  l = []
  le = len(s)
  for i in range(le):
    p=s[i]
    c = i+1
    while(c <le):
      q = s[c]
      p +=q
      l.append(p)
      c +=1
  #print(l)
  d = {}
  for i in l:
    if i not in d:
      d[i] = 1
    else:
      d[i] = d[i]+1
  #print(d)
  l1 = []
  for i,j in d.items():
    if j == 2 and len(i) == 3:
      l1.append(j)
  if len(l1) != 0:
    return("Yes")
  else:
    return("No")

def utility(n):
  
  z = zeros(n)      #checking if zeroes follow first digit or not
  s = allSame(n)    #checking if all digits are same or not
  con = consecuitive(n) #checking if consecutive 
  p = palindrome(n) #checking for palindrome
  u = unique(n)     #checking for unique numbers
  
  if z == 'Yes' or s == 'Yes' or con == 'Yes' or p == 'Yes' or u == 'Yes':
    return("Yes")
  else:
    return("No")


n = input()
print(utility(n))
```

```python
class House:
  def __init__(self,w,d):
    self.windows = w
    self.doors = d
  
  def view(self):
    print("The house has {} windows and {} doors".format(self.windows, self.doors))

  def __add__(self,other):
    new_windows = self.windows+other.windows 
    new_doors = self.doors+other.doors
    obj = House(new_windows,new_doors)
    return obj

h1 = House(6,2)
h2 = House(4,1)
h3 = h1+h2 #h1.__add__other
h1.view()
h2.view()
h3.view()
```

```python
from abc import ABC, abstractmethod

#Employee class
class Employee(ABC):
	
	#abstract method
	@abstractmethod
	def monthly_pay(self):
		pass

class Hourly_Employee(Employee):
	
	hourly_earnings = 10

	def monthly_pay(self):
		return Hourly_Employee.hourly_earnings + 4*30

	def annual_pay(self):
		return Hourly_Employee.hourly_earnings + 4*30*12

class Salaried_Employee(Employee):
	
	monthly_earnings = 20000

	def monthly_pay(self):
		return Salaried_Employee.monthly_earnings

	def annual_pay(self):
		return Salaried_Employee.monthly_earnings*12


david = Hourly_Employee()

print("David's monthly pay is:",david.monthly_pay())
print("David's annual pay is:",david.annual_pay())

print("-------------------------------------------")

sarah = Salaried_Employee()

print("Sarah's monthly pay is:",sarah.monthly_pay())
print("Sarah's annual pay is:",sarah.annual_pay())
```
