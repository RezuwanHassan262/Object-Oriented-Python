```python
class Customer:

  def __init__(self,name,*items):
    self.name = name
    self.items = items

  def purchase(self, *items):
    l = len(items) 
    print(self.name, ', you purchased', l,'item(s):')
    for i in items:
      print(i)

  def greet(self, name = None):
    if name == None:
      print("Hello!")
    else:
      print("Hello!", self.name)


customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
class Panda:

  def __init__(self,name,gender,age):
    self.name = name
    self.gender = gender
    self.age = age

  def sleep(self,x = None):
    if x != None:
      if 3<=x<=5:
        meals = "Mixed Veggies"
      elif x<=8:
        meals = "Eggplant & Tofu"
      elif x<=11:
        meals = "Broccoli Chicken"
      return("{} sleeps {} hours daily and should have {}.".format(self.name,x,meals))
    else:
      meals = "Bamboo Leaves"
      return("{}'s duration is unknown thus should have only {}.".format(self.name,meals))


panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
class Student:
  def __init__(self,name=None):
    if name != None:
      self.name = name
    else:
      self.name = "Default student"


  def quizcalc(self,*nums):
    if len(nums) != 1:
      sum = 0
      for i in nums:
        sum += i
      self.avg = sum/3
    else:
      self.avg = nums[0]/3


  def printdetail(self):
    print("Hello {}!".format(self.name))
    print("Your average quiz score is {}".format(self.avg))


s1 = Student()
s1.quizcalc(10)
print('--------------------------------')
s1.printdetail()
s2 = Student('Harry')
s2.quizcalc(10,8)
print('--------------------------------')
s2.printdetail()
s3 = Student('Hermione')
s3.quizcalc(10,9,10)
print('--------------------------------')
s3.printdetail()
```

```python
class Vehicle:
  def __init__(self):
    self.ip = [0,0]
    
  def moveUp(self):  
    self.ip[1] = self.ip[1]+1

  def moveDown(self):  
    self.ip[1] = self.ip[1]-1

  def moveLeft(self):  
    self.ip[0] = self.ip[0]-1

  def moveRight(self):
    self.ip[0] = self.ip[0]+1
    

  def print_position(self):
    print(tuple(self.ip)) 


car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
class Programmer:
  def __init__(self,name,lang,exp):
    self.name = name
    self.lang = lang
    self.exp = exp
    x = exp
  
  def printDetails(self):
    if self.exp == x:
      print("Horray! A new programmer is born")
      print("Name: {}".format(self.name))
      print("Language: {}".format(self.lang))
      print("Experience: {} years".format(self.exp))
    else:
      print("Updating experience of {}".format(self.name))
      print("Name: {}".format(self.name))
      print("Language: {}".format(self.lang))
      print("Experience: {} years".format(self.exp))  

  def addExp(self, new_exp):
    self.exp = self.exp+new_exp

p1 = Programmer("Ethen Hunt", "Java", 10)
p1.printDetails()
print('--------------------------')
p2 = Programmer("James Bond", "C++", 7)
p2.printDetails()
print('--------------------------')
p3 = Programmer("Jon Snow", "Python", 4)
p3.printDetails()
p3.addExp(5)
p3.printDetails()
```

```python
class ParcelKoro:
  
  def __init__(self, name = "No name set", weight=0):
    self.name = name
    self.weight = weight


  def product_weight(self,pw):
    self.weight = pw

  def calculateFee(self,place = 'default'):
    if place == 'default' and self.weight != 0:
      self.tf = (self.weight*20) + 50
    elif place != 'default' and self.weight != 0:
      self.tf = (self.weight*20) + 100  
    else:
      self.tf = 0


  def printDetails(self):
    print("Customer Name: {}".format(self.name))
    print("Product Weight: {}".format(self.weight))
    print("Total Fee: {}".format(self.tf))



print("**********************")
p1 = ParcelKoro()
p1.calculateFee()
p1.printDetails()
print("**********************")
p2 = ParcelKoro('Bob The Builder')
p2.calculateFee()
p2.printDetails()
print("----------------------------")
p2.product_weight(15)
p2.calculateFee()
p2.printDetails()
print("**********************")
p3 = ParcelKoro('Dora The Explorer', 10)
p3.calculateFee('Dhanmondi')
p3.printDetails()
```

```python
class Batsman:

  def __init__(self, *items):
    if len(items) == 3:
      self.name = items[0]
      self.runScored = items[1]
      self.ballsFaced = items[2]
    else:
      self.name = "New Batsman"
      self.runScored = items[0]
      self.ballsFaced = items[1]

    self.bsr = 0

  def printCareerStatistics(self):
    if self.name != "New Batsman":
      print("Name: {}".format(self.name))
      print("Runs Scored: {}, Balls Faced: {}".format(self.runScored,self.ballsFaced))
    else:
      print("Name: {}".format(self.name))
      print("Runs Scored: {}, Balls Faced: {}".format(self.runScored,self.ballsFaced))

  def setName(self,newName):
      self.name = newName

  def battingStrikeRate(self):
      self.bsr = (self.runScored/self.ballsFaced)*100
      return(self.bsr)



b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
class Account:
  def __init__(self,name = "Default Account",deposit = 0.0):
    self.name = name
    self.deposit = deposit
    self.balance = 0.0

  def name(self,NewName):
    self.name = NewName

  def details(self):
    print(self.name)
    self.balance = self.balance + self.deposit
    return(self.balance)

  def withdraw(self,NewWithdraw):
    x = self.balance - NewWithdraw 
    if x <= 3070:
      print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
    else:
      self.balance = self.balance - NewWithdraw 
      print("Withdraw successful! New balance is: {}".format(self.balance))



a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930);
print("------------------------")
a2.withdraw(600);
print("------------------------")
a1.withdraw(6929)
```

```python
class Author:
  
  def __init__(self, name = "Default", *bookName):
    self.name = name
    self.bookName = bookName
    self.books = bookName

    if self.books != " ":
      s = ""
      for i in self.books:
        s = s+i+"\n"
      self.books = s



  def addBooks(self,*newBookList):
    l = list(newBookList)
    s = ""
    for i in l:
      s = s+i+"\n"
    self.books = s



  def printDetails(self):
    print("Author Name: {}".format(self.name))
    print("--------")
    print("List of Books: ")  
    print(self.books)

  def changeName(self,newName):
    self.name = newName

auth1 = Author('Humayun Ahmed')
auth1.addBooks('Deyal', 'Megher Opor Bari')
auth1.printDetails()
print("===================")
auth2 = Author()
print(auth2.name)
auth2.changeName('Mario Puzo')
auth2.addBooks('The Godfather', 'Omerta', 'The Sicilian')
print("===================")
auth2.printDetails()
print("===================")
auth3 = Author('Paolo Coelho', 'The Alchemist', 'The Fifth Mountain')
auth3.printDetails()
```
