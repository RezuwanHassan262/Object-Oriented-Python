```python
dictio = {}

s = input()

combs = []
for turn in range(1, len(s)+1):
    index = 0
    while index <= len(s) - turn:
        new = s[index: index+turn]
        combs.append(new)
        index += 1

for item in combs:
    if 'A' in item and 'B' in item:
        first = item[0]
        index = None

        for index in range(1, len(item)):
            if item[index] != first:
                break

        if first not in item[index: ]:
            if item.count('A') == item.count('B'):
                if item not in dictio.keys():
                    dictio.update({item: 1})
                else:
                    dictio[item] += 1

for key, val in dictio.items():
    print(f"{key} - {val} times")

patterns = list(dictio.keys())
patterns.sort(key = len)

same_lens = [patterns[-1]]
for index in range(len(patterns)-1, 0, -1):
    if len(patterns[index]) == len(patterns[index-1]):
        same_lens.append(patterns[index-1])
    else:
        break

same_lens.sort()

print(f"Longest Pattern - {same_lens[0]}")
```
