```python
n = int(input("How many players will be there in total?: "))
l = []
for i in range(n):
  i = input('''Insert player details in such format: "Year of winning" "Tournament name" "Player name": ''')
  i = i.split()
  l.append(i)
#print(l)



dj = []
na = []
do = []
for i in l:
  if i[2]=="Djokovic":
    dj.append(i)
  elif i[2]=="Nadal":
    na.append(i)
  else:
    do.append(i)

#print(dj)
#print(na)
#print(do)



d1 = {}
y2021 = []
y2020 = []
y2019 = []

for i in dj:
  if i[0] == '2021':
    y2021.append(i[1])
  elif i[0] == '2020':
    y2020.append(i[1])
  else:
    y2019.append(i[1])

if any(y2021) != True:
  pass
else:
  d1["2021"] = y2021

if any(y2020) != True:
  pass
else:
  d1["2020"] = y2020

if any(y2019) != True:
  pass
else:
  d1["2019"] = y2019


d2 = {}
y2021 = []
y2020 = []
y2019 = []

for i in na:
  if i[0] == '2021':
    y2021.append(i[1])
  elif i[0] == '2020':
    y2020.append(i[1])
  else:
    y2019.append(i[1])

if any(y2021) != True:
  pass
else:
  d2["2021"] = y2021

if any(y2020) != True:
  pass
else:
  d2["2020"] = y2020

if any(y2019) != True:
  pass
else:
  d2["2019"] = y2019


d3 = {}
y2021 = []
y2020 = []
y2019 = []

for i in do:
  if i[0] == '2021':
    y2021.append(i[1])
  elif i[0] == '2020':
    y2020.append(i[1])
  else:
    y2019.append(i[1])

if any(y2021) != True:
  pass
else:
  d3["2021"] = y2021

if any(y2020) != True:
  pass
else:
  d3["2020"] = y2020

if any(y2019) != True:
  pass
else:
  d3["2019"] = y2019

d = {}
d["Djokovic"] = d1
d["Nadal"] = d2
d["Dominic"] = d3
print('\n')
print(d)
```
