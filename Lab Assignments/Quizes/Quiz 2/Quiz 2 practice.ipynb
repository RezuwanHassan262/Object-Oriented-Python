```python
class Student:
  def __init__(self,name = None,dept = None):
    self.name = name
    self.dept = dept
    self.courses = None

    if self.name == None and self.dept == None:
      print("Student name and department need to be set")
    elif self.name != None and self.dept == None:
      print("Department for {} needs to be set".format(self.name))
    else:
      print("{} if from {} department".format(self.name,self.dept))


  def update_name(self,newName):
    self.name = newName

  def update_department(self,newDept):
    self.dept = newDept

  def update_courses(self,*newCourses):
    self.courses = newCourses

  def enroll(self,*courses):
    self.courses = courses

  def printDetail(self):
    print("Name: {}".format(self.name))
    print("Department: {}".format(self.dept))
    self.l = len(self.courses)
    s = ""
    for i in self.courses:
      s = s+i+", "
    s = s[0:(len(s)-2)]
    print("{} enrolled in {} courses(s)".format(self.name,self.l))
    print(s)



s1=Student()
print("=========================")
s2 = Student("Carol")
print("=========================")
s3 = Student("Jon", "EEE")
print("=========================")
s1.update_name("Bob")
s1.update_department("CSE")
s2.update_department("BBA")
s1.enroll("CSE110", "MAT110", "ENG091")
s2.enroll("BUS101")
s3.enroll("MAT110", "PHY111")
print("###########################")
s1.printDetail()
print("=========================")
s2.printDetail()
print("=========================")
s3.printDetail()
```

```python
class Student:
  def __init__(self,name,id,dept):
    self.name = name
    self.id = id
    self.dept = dept
  
  def details(self):
    x = "Name: {}".format(self.name)
    y = "ID: {}".format(self.id)
    z = "Department: {}".format(self.dept)
    x1 = x+'\n'+y+'\n'+z
    return x1

  def advise(self,*courses):
    l = list(courses)
    l1 = len(l)
    s = ""
    for i in l:
      s = s+i+", "
    s = s[0:(len(s)-2)] 
    c = l1*3.0
    if 9.0<=c<=12.0:
      print("{}, you have taken {} credits".format(self.name,c))
      print("List of courses: {}".format(s))
      print("Status: Ok")
    elif c < 9.0:
      print("{} you have taken {} credits".format(self.name,c))
      print("List of courses: {}".format(s))
      print("Status: You have to take at least 1 more course")
    else:
      print("{} you have taken {} credits".format(self.name,c))
      print("List of courses: {}".format(s))
      print("Status: You have to drop at least 1 more course")




s1 = Student("Alice","20103012","CSE")
s2 = Student("Bob", "18301254","EEE")
s3 = Student("Carol", "17101238","CSE")
print("##########################")
print(s1.details())
print("##########################")
print(s2.details())
print("##########################")
s1.advise("CSE110", "MAT110", "PHY111")
print("##########################")
s2.advise("BUS101", "MAT120")
print("##########################")
s3.advise("MAT110", "PHY111", "ENG102", "CSE111", "CSE230")
```

```python
class Author:

  def __init__(self,name = None):
    self.name = name
    self.horror = []
    self.sf = []

  def addBook(self,book,genre):

    if self.name == None:
      print("A book can not be set without author name")
    else:
      if genre=="Horror":
        self.horror.append(book)
      else:
        self.sf.append(book)
      

  def setName(self,name):
    self.name = name 

  def printDetail(self):
    l = len(self.sf)+len(self.horror)
    
    scf = ""
    for i in self.sf:
      scf = scf+i+", "
    scf = scf[0:(len(scf)-2)]
    
    hr = ""
    for i in self.horror:
      hr = hr+i+", "
    hr = hr[0:(len(hr)-2)]


    if len(self.sf) != 0 and len(self.horror) == 0:
      x = '''Number of Book(s): {} \nAuthor Name: {} \nScience Fiction: {}'''.format(l,self.name,scf)
    elif len(self.sf) == 0 and len(self.horror) != 0:
      x = '''Number of Book(s): {} \nAuthor Name: {} \nHorror: {}'''.format(l,self.name,hr)
    elif len(self.sf) != 0 and len(self.horror) != 0:
      x = '''Number of Book(s): {} \nAuthor Name: {} \nScience Fiction: {} \nHorror: {}'''.format(l,self.name,scf,hr)
    
    print(x)



a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```

```python
class Vaccine:
  def __init__ (self,name, origin,frequency):
    self.name = name
    self.origin = origin
    self.frequency = frequency

class Person:
  def __init__(self,name,age,profession = None):
    self.name = name
    self.age = age
    self.profession = profession
    self.vaccine = None
    self.firstDose = None
    self.secondDose = None
  
  def pushVaccine(self,vac,curDose = None):
    if self.profession == 'Student' or self.age >=25:
      if curDose == None:
        self.dose1 = "Given"
        self.dose2 = "Please come back after {} days".format(vac.frequency)
        self.vaccine = vac
        print("First dose done for {}".format(self.name))
      else:
        if self.vaccine.name == vac.name:
          self.secondDose == "Given"
          print("Second dose complete for {}".format(self.name))
        else:
          print("Sorry {}, You can't have 2 types of vaccine".format(self.name))
    else:
      print("sorry {}, you need to be atleast 25 to take a vaccine".format(self.name))

  def showDetail(self):
    x = "Name: {} Age: {} Type: {} \nVaccine name: {} \n1st dose: {} \n2nd dose: {}".format(self.name,self.age,self.profession,self.vaccine.name,self.dose1,self.dose2)
    print(x)


if __name__ == "__main__":
    astra = Vaccine("AstraZeneca", "UK", 60)
    modr = Vaccine("Moderna", "UK", 30)
    sin = Vaccine("Sinopharm", "China", 30)
    p1 = Person("Bob", 21, "Student")
    print("=================================")
    p1.pushVaccine(astra)
    print("=================================")
    p1.showDetail()
    print("=================================")
    p1.pushVaccine(sin, "2nd Dose")
    print("=================================")
    p1.pushVaccine(astra, "2nd Dose")
    print("=================================")
    p1.showDetail()
    print("=================================")
    p2 = Person("Carol", 23, "Actor")
    print("=================================")
    p2.pushVaccine(sin)
    print("=================================")
    p3 = Person("David", 34)
    print("=================================")
    p3.pushVaccine(modr)
    print("=================================")
    p3.showDetail()
    print("=================================")
    p3.pushVaccine(modr, "2nd Dose")
```
