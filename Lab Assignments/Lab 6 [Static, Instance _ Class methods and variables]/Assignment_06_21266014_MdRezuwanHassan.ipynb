```python
class Assassin:

  count = 0
  def __init__(self,name,sRate):
    self.name = name
    self.sRate = sRate
    Assassin.count += 1
  
  def printDetails(self):
    print("Name:", self.name)
    print("Succss Rate: {}%".format(self.sRate))
    print("Total number of Assassin:", Assassin.count)

  @classmethod
  def failureRate(cls,name, fRate):
    sRate = 100 - fRate
    return cls(name,sRate)

  @classmethod
  def failurePercentage(cls,name, fRate):
    sRate = 100 - fRate
    return cls(name,sRate)


john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
class Passenger:
  count = 0
  def __init__(self,name):
    self.name = name
    self.bFare = 450
    Passenger.count += 1
  
  def set_bag_weight(self,weight):
    if weight <= 20:
      extra = 0
    elif weight <= 50:
      extra = 50
    else:
      extra = 100
    self.tFare = self.bFare + extra
  
  def printDetail(self):
    print("Name: {}".format(self.name))
    print("Bus Fare: {}".format(self.tFare))
    

print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
class Employee:

  def __init__(self,name,eYear):
    self.name = name
    self.workingPeriod = eYear
  
  @classmethod
  def employeeByJoiningYear(cls,name,jYear):
    eYear = 2021-jYear
    return cls(name,eYear)
  
  @classmethod
  def experienceCheck(cls,exp,gen):
    if exp >= 3:
      if gen == "male":
        return "He is experienced"
      else:
        return "He is not experienced"
    else:
      if gen == "female":
        return "She is experienced"
      else:
        return "She is not experienced"



employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
class Laptop:
  laptopCount = 0
  def __init__(self,name,count):
    self.name = name
    self.count = count
    Laptop.laptopCount += count

  @staticmethod
  def advantage():
    print("Laptops are portable")

  @staticmethod
  def resetCount():
    Laptop.laptopCount = 0

lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
class Cat:
  Number_of_cats = 0

  def __init__(self,color,action):
    self.color = color
    self.action = action
    Cat.Number_of_cats += 1

  def printCat(self):
    print("{} cat is {}".format(self.color,self.action))

  @classmethod
  def no_parameter(cls):
    return cls("White","sitting")

  @classmethod
  def first_parameter(cls,color,):
    return cls(color,"sitting")

  @classmethod
  def second_parameter(cls,action,):
    return cls("Grey",action)

  def changeColor(self,nCol):
    self.color = nCol

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
import math
class Cylinder:
  radius = 5
  height = 18
  
  def __init__(self,r,h):
    print('Default: Radius = {} and height = {}'.format(Cylinder.radius,Cylinder.height))
    self.radius = float(r)
    self.height = float(h)
    Cylinder.radius = float(r)
    Cylinder.height = float(h)
    print('Updated: Radius = {} and height = {}'.format(self.radius,self.height))

  @staticmethod
  def area(radius, height):
    arr = (2*math.pi*radius**2) + (2*math.pi*radius*height)
    print("Area: {}".format(arr))

  @staticmethod
  def volume(radius, height):
    vol = math.pi*(radius**2)*height
    print("Area: {}".format(vol))

  @classmethod
  def swap(cls,height,radius):
    return cls(radius,height)

  @classmethod
  def changeFormat(cls,info):
    rad, hei = info.split("-")
    return cls(rad,hei)


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
class SultansDine:
  totalBranches = 0
  totalSell = 0
  branch = list()

  def __init__ (self,branch):
    self.branch = branch
    self.Branch_sell = 0
    self.amount = 0
    SultansDine.totalBranches += 1
    SultansDine.branch.append(self)

  @classmethod
  def details(cls):
    print("Total Number of branch(s):",cls.totalBranches)    
    print("Total Sell::",cls.totalSell)
    for i in SultansDine.branch:
      per = (i.Branch_sell/SultansDine.totalSell)*100
      print("Branch Name: {}, Branch Sell: {} Taka".format(i.branch, i.Branch_sell))
      print("Branch consists of total sell's: {}%".format(per))

  def sellQuantity(self,amount):
    if amount <10:
      Branch_sell = amount * 300   
    elif amount <20:
      Branch_sell = amount * 350   
    else:
      Branch_sell = amount * 400   

    self.Branch_sell = Branch_sell
    SultansDine.totalSell += Branch_sell 

  def branchInformation(self):
    print("Branch Name:",self.branch)    
    print("Branch Sell::",self.Branch_sell)


#################################


SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
